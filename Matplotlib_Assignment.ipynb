# Scipy

We have the min and max temperatures in a city In India for each months of the year.<br>
We would like to find a function to describe this and show it graphically, the dataset given below.<br>
Task:
1. fitting it to the periodic function
2. plot the fit <br>
   <b>Data</b><br>
   Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25<br>
   Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize

temp_max = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
temp_min = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])

months = np.arange(12)

def yearly_temperatures(times, avg, ampl, time_offset):
    return (avg
            + ampl * np.cos((times + time_offset) * 1.8 * np.pi / times.max()))

res_max, cov_max = optimize.curve_fit(yearly_temperatures, months,
                                      temp_max, [40, 20, 0])
res_min, cov_min = optimize.curve_fit(yearly_temperatures, months,
                                      temp_min, [-40, 20, 0])
```

```python
days = np.linspace(0, 12, 365)

plt.figure()
plt.plot(months, temp_max, 'ro')
plt.plot(days, yearly_temperatures(days, *res_max), 'r-')
plt.plot(months, temp_min, 'bo')
plt.plot(days, yearly_temperatures(days, *res_min), 'b-')
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')

plt.show()
```

# Matplotlib

This assignment is for visualization using matplotlib:
data to use:<br>
url=https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv<br>
titanic = pd.read_csv(url)
Charts to plot:
1. Create a pie chart presenting the male/female proportion
2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
import pandas as pd
import matplotlib.pyplot as plt

url='https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'

titanic=pd.read_csv(url,skip_blank_lines=True)

count=titanic['sex'].value_counts()
colors = ['green', 'yellow']
plt.pie(count, labels=count.index, colors=colors, autopct='%1.1f%%')
plt.title('Male / Female \nProportion')
```

```python
malecolor={'male' : 'blue'}
femalecolor={'female': 'red'}
plt.scatter(titanic[titanic['sex']=='male']['age'], titanic[titanic['sex']=='male']['fare'],  c=titanic[titanic['sex']=='male']['sex'].map(malecolor), label='Male')
plt.scatter(titanic[titanic['sex']=='female']['age'], titanic[titanic['sex']=='female']['fare'],  c=titanic[titanic['sex']=='female']['sex'].map(femalecolor), label='Female')
plt.legend()
plt.title('Fare paid / Age by Gender')
plt.xlabel('Age')
plt.ylabel('Fare')
```
